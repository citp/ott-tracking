```python
import json
from os.path import basename, join

EOL = "\n"


def convert_unmitmable(filename, outdir):
    tuples = set()
    for l in open(filename):
        l = l.rstrip()
        if not l:
            continue
        if "\t" in l:
            parts = l.split("\t")
            if len(parts) == 3:
                ch_id, ip, domain = parts
            elif len(parts) == 2:
                ch_id, ip = parts
                domain = ""
        else:
            parts = l.split(" ", 2)
            ch_id = parts[1].replace('"', '').strip(":")
            b = json.loads(parts[2])
            assert len(b) == 1
            domain = b.keys()[0]
            ip_port = b.values()[0]
            ip = ip_port.split(" ")[0].split("'")[1]
            # port = ip_port.split(" ")[1].rstrip(")")
            if not ip:
                raise ValueError("convert_unmitmable: No IP address")
        tuples.add((ch_id, ip, domain))
    f_basename = basename(filename)
    f_out = "%s.unmitmable" % f_basename.split("-")[0]
    with open(join(outdir, f_out), "w") as f:
        for _tuple in tuples:
            f.write("\t".join(_tuple) + EOL)
```

```python
from glob import glob
# unmitmable directory contains *.unmitmable files from the 20190131-112137/mitmlogs folder
for f in glob("/tmp/unmitmable/*"):
    convert_unmitmable(f, "/tmp/unmitmable_out")
```

```python
from os.path import isfile
def loadUnMitmableHostsAndIps(filename):
    hosts = set()
    ips = set()
    if isfile(filename):
        for line in open(filename):
            line = line.rstrip("\n")
            _, ip, host = line.split("\t")
            if host:
                hosts.add(host)
            if ip:
                ips.add(ip)
    return hosts, ips
```
